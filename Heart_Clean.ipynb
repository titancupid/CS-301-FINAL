```python
import pandas as pd

df = pd.read_csv('heart.csv')

#df.head()

#df.isnull().sum()

'''
Age--Numerical, Discrete
Sex--Categorical, Nominal
ChestPainType--Categorial, Nominal
RestingBP--Numerical, Discrete
Cholesterol--Numerical, Discrete
FastingBS--Numerical, Discrete
RestingECG--Categorical, Nominal
MaxHR--Numerical, Discrete
ExerciseAngina--Categorical, Nominal
Oldpeak--Numerical, Continuous
ST_Slope--Categorical, Nominal
HeartDisease--Categorical, Nominal
'''
#number_of_missing_values=df.isnull().sum()
#Confirmation for null values
#print("Total missing values in dataset:", number_of_missing_values.sum())

# Number of duplicate rows
#print(df.duplicated().sum())


'''
Age                 int64
Sex                object
ChestPainType      object
RestingBP           int64
Cholesterol         int64
FastingBS           int64
RestingECG         object
MaxHR               int64
ExerciseAngina     object
Oldpeak           float64
ST_Slope           object
HeartDisease        int64
dtype: object
'''




#Delete a '0' resting BP row
zero_restingBP = df['RestingBP'] == 0
selectedR = df[zero_restingBP]
df.drop(selectedR.index, inplace=True)

#Fill rows with '0' cholesterol with the median cholesterol


zero_cholesterol = df['Cholesterol'] == 0

df.loc[zero_cholesterol, 'Cholesterol'] = df['Cholesterol'].median()


selectedC = df[zero_cholesterol]
print(selectedC)

import pandas as pd

#One-Hot Encoding
one_hot_encoded_df = pd.get_dummies(df,columns=['Sex','ChestPainType','RestingECG','ExerciseAngina', 'ST_Slope', 'HeartDisease'], drop_first=False).astype(float)
print(one_hot_encoded_df)

one_hot_encoded_df.describe()


from google.colab import files
one_hot_encoded_df.to_csv('heart_clean.csv', index=False)

#files.download('heart_clean.csv')
```
