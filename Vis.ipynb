```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn as skl
```

```python
orgDF=pd.read_csv("heart.csv")

DF=pd.read_csv("heart_clean.csv")
clDF=DF.drop(['HeartDisease_0', "Sex_M", "ExerciseAngina_N"], axis =1)
clDF.info()

numDF=clDF[["Age","RestingBP","Cholesterol","Oldpeak", "MaxHR","Sex_F"]]
catDF=clDF[["Sex_F", "ChestPainType_ASY", "ChestPainType_ATA","ChestPainType_NAP","ChestPainType_TA","RestingECG_Normal","RestingECG_ST","ExerciseAngina_Y", "ST_Slope_Down","ST_Slope_Flat","ST_Slope_Up", "HeartDisease_1"]]
numDF.head()
```
